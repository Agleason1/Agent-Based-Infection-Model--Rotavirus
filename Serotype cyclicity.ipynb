```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv('serotype_prevalence_by_year.csv')
df
```

```python
all_serotypes=np.unique(df['serotype'])

serotype_prev_dict=dict()

for serotype in all_serotypes:
    serotype_prev_dict[serotype]=[]


for year in range(2012,2021):
    df_now=df[df['year']==year]
    serotypes_now=df_now['serotype']
    print(year)
    prevalences_now=[df_now['proportion'].values[0]]+np.diff(df_now['proportion']).tolist()
    sum_prev_now=sum(prevalences_now)
    prevalences_now=[i/sum_prev_now for i in prevalences_now]
    
    serotype_prev_now_dict=dict(zip(serotypes_now,prevalences_now))
    
    for serotype in all_serotypes:
        if serotype not in serotypes_now.values:
                serotype_prev_dict[serotype]+=[0]
        else:
            serotype_prev_dict[serotype]+=[serotype_prev_now_dict[serotype]]
    
            
    
    
    
```

```python
# fit a line to the economic data
from numpy import sin,cos
from numpy import sqrt
from numpy import arange
from pandas import read_csv
from scipy.optimize import curve_fit
from matplotlib import pyplot
from scipy.optimize import curve_fit
```

```python
# define the true objective function
def objective(x, a, b, c, d):
    return a * sin(b - x) + c * x + d
 
```

```python
serotypes=[]
best_parameters=[]
formula=[]

for curve,serotype in zip(serotype_prev_dict.values(),serotype_prev_dict.keys()):

    # choose the input and output variables
    x, y = np.arange(0,len(curve)), curve
    # curve fit
    popt, _ = curve_fit(objective, x, y)
    # summarize the parameter values
    a, b, c, d = popt
    print(popt)
    # plot input vs output
    pyplot.scatter(x, y)
    # define a sequence of inputs between the smallest and largest known inputs
    x_line = np.linspace(min(x), max(x), 2000)
    # calculate the output for the range
    y_line = objective(x_line, a, b, c, d)
    
    serotypes+=[serotype]
    best_parameters+=[[a,b,c,d]]
    formula+=['a * sin(b - x) + c * x**2 + d']
    
    
    # create a line plot for the mapping function
    pyplot.plot(x_line, y_line, '--', color='red')
    pyplot.show()
```
